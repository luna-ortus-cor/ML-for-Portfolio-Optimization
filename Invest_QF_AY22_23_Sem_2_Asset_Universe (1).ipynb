```python
def get_variable_name(variable):
    globals_dict = globals()

    return [var_name for var_name in globals_dict if globals_dict[var_name] is variable]

def getstockdata(assetlist):
  for sector in assetlist:
    sectorname = get_variable_name(sector)[0]
    for stock in sector[:5]:
      df = yf.download(tickers = stock, start="2018-01-01", end="2022-12-31", interval = '1d', group_by='ticker', threads=True)
      df.drop(columns=[lst for lst in df.columns if "Volume" in lst or ("Close" in lst and not "Adj Close" in lst)], inplace=True)
      df.to_csv('/content/' + str(sectorname) + '/' + str(stock) + '.csv')
    
getstockdata(assets)
```

```python
import pandas as pd
import numpy as np
import os
from scipy.stats import norm
```

```python
path = 'C:\\Users\\13643\\OneDrive - National University of Singapore\\Desktop\\School\\CCAs\\NUS Investment Society QF\\Final Projects\\ML-for-Portfolio-Optimization-main\\data'
os.chdir(path)
industry_name = os.listdir(path)

def get_industry_price(industry):
    opt = []
    temp_path = path + '\\' + industry
    os.chdir(temp_path)
    temp_file_name = os.listdir(temp_path)
    opt = pd.DataFrame(pd.read_csv(temp_file_name[0])['Date'])
    for stock in temp_file_name:
        opt[industry + '_' + stock.rstrip('.csv')] = pd.Series(pd.read_csv(stock).iloc[:,4])
    return opt
        
price_matrix = get_industry_price(industry_name[0]).iloc[:,1:]
temp_matrix = [get_industry_price(industry) for industry in industry_name[1:]]
for ind_pd in temp_matrix:
    for stock in ind_pd.columns[1:]:
        price_matrix[stock] = ind_pd[stock]
price_matrix
```

**Initialize**

```python
def get_alpha(x,beta=0.05):
    temp_mu = x.mean()
    temp_sigma = x.std()
    temp_alpha = norm.ppf(beta, temp_mu, temp_sigma**(1/2))
    return temp_alpha

weight_vector = pd.DataFrame([1/len(X) for i in range(len(X))])
wX = weight_vector.values*X
alpha = [get_alpha(wX.iloc[i]) for i in range(len(wX))]
z = [-item if item < 0 else 0 for item in alpha]
```
